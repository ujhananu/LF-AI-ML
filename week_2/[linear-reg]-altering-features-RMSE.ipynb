### Split the dataset into train set and test set.

`train_test_split()` accepts `test_size` parameter that specify the test set proportion. `.2` means `20%` of the total data is split as test set.

`random_state` parameters helps to create re-producible train and test split. On each invocation of `train_test_split()` with `random_state=42` creates the same train and test split. See more here: https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=.2, random_state=42)
```

```python
print(f"Train Shape: {train.shape}")
print(f"Test Shape: {test.shape}")
```

Separate out the feature and target column in train and test set.

```python
train_x = train[feature_column]
train_y = train[target_column]

test_x = test[feature_column]
test_y = test[target_column]
```

## Linear Regression model 

Import LinearRegression model

```python
from sklearn.linear_model import LinearRegression


model = LinearRegression()
```

Call `.fit()` method to learn the model parameters. 

In $y = wx + b$, `w` and `b` are the model parameters. `.fit()` computes these parameters.

```python
model.fit(train_x, train_y)
```

Print learned `w` here.

See how no. of values change on changing the no. of feature columns.

```python
model.coef_
```

Print bias/intercept `b` here.

```python
model.intercept_
```

Use the learned linear regression model to predict on the test data set.

```python
y_pred = model.predict(test_x)
```

```python
print(y_pred)
```

Import `root_mean_squared_error` from `sklearn.metrics` and calculate the RMSE for the predictions.

https://scikit-learn.org/stable/modules/generated/sklearn.metrics.root_mean_squared_error.html

```python
print(df['SalePrice'].mean())
```

```python
# write your code here
from sklearn.metrics import mean_squared_error, root_mean_squared_error


MSE = mean_squared_error(test_y, y_pred)
print(f"Mean Squared Error: {MSE}")

# calculate RMSE
RMSE = np.sqrt(mean_squared_error(test_y, y_pred))
print(f"Root Mean Squared Error: {RMSE}")
```

```python
rmse_percent = RMSE / df['SalePrice'].mean() * 100
print(f"RMSE % of average price: {rmse_percent:.2f}%")
```

RMSE is around 19% , which is lesser than the previous one i.e 33%. This shows that this model is better
